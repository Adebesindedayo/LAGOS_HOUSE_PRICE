```python
def _85th(x):
    return x.quantile(0.90)
def _15th(x):
    return x.quantile(0.10)
```

```python
df['_85th'] = df.groupby(['property_area', 'bed'])['price'].transform(_90th)
df['_15th'] = df.groupby(['property_area', 'bed'])['price'].transform(_10th)
df.head()
```

```python
df.shape
```

```python
df.drop(df[(df['price'] > df['_85th']) | (df['price'] < df['_15th'])].index, inplace=True)
```

```python
df.shape
```

```python
df.drop(["_85th", "_15th"], axis=1, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df["rank"] = df.groupby(['specific_location'])['price'].rank(method='dense')
df
```

```python
df = df.sort_values('rank')
df
```

```python
df1 = df.sort_values('price')
df1
```

```python
labelencoder = LabelEncoder()
df1["label_df"] = labelencoder.fit_transform(df1['specific_location'])
```

```python
df1
```

```python
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
from math import sqrt
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import roc_curve,auc
from sklearn.metrics import r2_score
```

```python
features = df1.loc[:, ["bed", "bath", "toilet", "label_df", "rank"]].copy()
target = df1.loc[:, "price"].copy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(np.array(features), np.array(target), test_size = 0.2, random_state = 0)
X_train
```

```python
regressor = RandomForestRegressor(n_estimators=100, random_state=0)
regressor.fit(X_train, y_train)
y_predict = regressor.predict(X_test)

print('mean absolute error: {:.2f}'.format(mean_absolute_error(y_test, y_predict)))
print('mean squared error: {:.2f}'.format(mean_squared_error(y_test, y_predict)))
print('root mean squared error: {:.2f}'.format(np.sqrt(mean_squared_error(y_test, y_predict))))
print('r2_score: {:.2f}'.format(r2_score(y_test, y_predict)))
```
